```python
# Dependencies and Setup
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
# mouse_drug_data_to_load = "data/mouse_drug_data.csv"
# clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

mouse_file = os.path.join("data", "mouse_drug_data.csv")
clinic_file = os.path.join("data","clinicaltrial_data.csv")


# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse_file)
clinic_df = pd.read_csv(clinic_file)

# Combine the data into a single dataset - Inner Join
merge_table = pd.merge(mouse_df, clinic_df, on="Mouse ID")

# Display the data table for preview
merge_table.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint
# Filtering merge_table
ShortTumor = merge_table[["Drug","Timepoint","Tumor Volume (mm3)"]]
GroupTumor = ShortTumor.groupby(["Drug","Timepoint"])


# Convert to DataFrame
Tumormean = GroupTumor.mean()
# Preview DataFrame
Tumorresetmean = Tumormean.reset_index()
Tumorresetmean
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

# Convert to DataFrame
TumorSterr = GroupTumor.sem()
# Preview DataFrame
Tumorresetsterr = TumorSterr.reset_index()
Tumorresetsterr
```

```python
# Minor Data Munging to Re-Format the Data Frames
Tumormeanpivot = Tumorresetmean.pivot(index="Timepoint", columns="Drug", values="Tumor Volume (mm3)")
# Preview that Reformatting worked
Tumormeanpivot
```

```python
#grouping drugs using .loc on Tumorresetmean
capomulin = Tumorresetmean.loc[Tumorresetmean["Drug"]=="Capomulin",:]
Infubinol = Tumorresetmean.loc[Tumorresetmean["Drug"]=="Infubinol",:]
Ketapril = Tumorresetmean.loc[Tumorresetmean["Drug"]=="Ketapril",:]
Placebo = Tumorresetmean.loc[Tumorresetmean["Drug"]=="Placebo",:]

#grouping drugs using.loc on Tumorresetsterr
capoerr = Tumorresetsterr.loc[Tumorresetsterr["Drug"]=="Capomulin",:]
Infuberr = Tumorresetsterr.loc[Tumorresetsterr["Drug"]=="Infubinol",:]
Ketaerr = Tumorresetsterr.loc[Tumorresetsterr["Drug"]=="Ketapril",:]
Placerr = Tumorresetsterr.loc[Tumorresetsterr["Drug"]=="Placebo",:]


# Generate the Plot (with Error Bars)
capomu = plt.errorbar(capomulin["Timepoint"],capomulin["Tumor Volume (mm3)"], yerr = capoerr["Tumor Volume (mm3)"], uplims=True, lolims=True, color="red",marker="o",label ="Capomulin")
Infub = plt.errorbar(Infubinol["Timepoint"],Infubinol["Tumor Volume (mm3)"], yerr = Infuberr["Tumor Volume (mm3)"], uplims=True, lolims=True,color="blue",marker="^",label ="Infubinol")
Keta = plt.errorbar(Ketapril["Timepoint"],Ketapril["Tumor Volume (mm3)"], yerr = Ketaerr["Tumor Volume (mm3)"], uplims=True, lolims=True,color="green",marker="s",label ="Ketapril")
plac = plt.errorbar(Placebo["Timepoint"],Placebo["Tumor Volume (mm3)"], yerr = Placerr["Tumor Volume (mm3)"], uplims=True, lolims=True,color="black",marker="d",label ="Placebo")

#Setting legend on best place in graph
plt.legend(loc="best")

plt.title("Tumor response to Treatment")
plt.xlabel("Time, Days")
plt.ylabel("Ave.Tumor Volume (mm3)")

# Save the Figure
plt.savefig("../Images/Treatment.png")

# Show the Figure
plt.show()
```

![Tumor Response to Treatment](../Images/treatment.png)

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint

# Filtering merge_table
ShortMeta = merge_table[["Drug","Timepoint","Metastatic Sites"]]
GroupMeta = ShortMeta.groupby(["Drug","Timepoint"])

# Convert to DataFrame
Metamean = GroupMeta.mean()
# Preview DataFrame
Metaresetmean = Metamean.reset_index()
Metaresetmean
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 

# Convert to DataFrame
MetaSterr = GroupMeta.sem()
# Preview DataFrame
Metaresetsterr = MetaSterr.reset_index()
Metaresetsterr
```

```python
# Minor Data Munging to Re-Format the Data Frames
Metameanpivot = Metaresetmean.pivot(index="Timepoint", columns="Drug", values="Metastatic Sites")
# Preview that Reformatting worked
Metameanpivot
```

```python
#grouping drugs using .loc on Metaresetmean
capomulin = Metaresetmean.loc[Metaresetmean["Drug"]=="Capomulin",:]
Infubinol = Metaresetmean.loc[Metaresetmean["Drug"]=="Infubinol",:]
Ketapril = Metaresetmean.loc[Metaresetmean["Drug"]=="Ketapril",:]
Placebo = Metaresetmean.loc[Metaresetmean["Drug"]=="Placebo",:]


# Generate the Plot (with Error Bars)
capomu = plt.plot(capomulin["Timepoint"],capomulin["Metastatic Sites"],color="red",marker="o",label ="Capomulin")
Infub = plt.plot(Infubinol["Timepoint"],Infubinol["Metastatic Sites"],color="blue",marker="^",label ="Infubinol")
Keta = plt.plot(Ketapril["Timepoint"],Ketapril["Metastatic Sites"],color="green",marker="s",label ="Ketapril")
plac = plt.plot(Placebo["Timepoint"],Placebo["Metastatic Sites"],color="black",marker="d",label ="Placebo")

#Setting legend on best place in graph
plt.legend(loc="best")

plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration, Days")
plt.ylabel("Metastatic Sites")

# Save the Figure
plt.savefig("../Images/Spread.png")

# Show the Figure
plt.show()
```

![Metastatic Spread During Treatment](../Images/spread.png)

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)

# Filtering merge_table
ShortMouse = merge_table[["Drug","Timepoint","Mouse ID"]]
GroupMouse = ShortMouse.groupby(["Drug","Timepoint"])

# Convert to DataFrame
Mousecount = GroupMouse.count()

# resetting dataframe Index
Mouseresetcount = Mousecount.reset_index()

#Re-naming the "Mouse ID" column
Mouseresetcount = Mouseresetcount.rename(columns={"Mouse ID":"Mouse Count"})

# Preview DataFrame
Mouseresetcount
```

```python
# Minor Data Munging to Re-Format the Data Frames
Mousecountpivot = Mouseresetcount.pivot(index="Timepoint", columns="Drug", values="Mouse Count")
# Preview that Reformatting worked
Mousecountpivot
```

```python
# Generate the Plot (Accounting for percentages)

#Maximum Mouse count for tested Drugs APPEARING IN PLOT COMPARISON
Max = Mousecountpivot["Capomulin"].max()
Mouseresetcount["Survival_Percent"] = Mouseresetcount["Mouse Count"]*100/Max
Mouseresetcount

#grouping drugs using .loc on Mouseresetcount
capomulin = Mouseresetcount.loc[Mouseresetcount["Drug"]=="Capomulin",:]
Infubinol = Mouseresetcount.loc[Mouseresetcount["Drug"]=="Infubinol",:]
Ketapril = Mouseresetcount.loc[Mouseresetcount["Drug"]=="Ketapril",:]
Placebo = Mouseresetcount.loc[Mouseresetcount["Drug"]=="Placebo",:]


# Generate the Plot (with Error Bars)
capomu = plt.plot(capomulin["Timepoint"],capomulin["Survival_Percent"],color="red",marker="o",label ="Capomulin")
Infub = plt.plot(Infubinol["Timepoint"],Infubinol["Survival_Percent"],color="blue",marker="^",label ="Infubinol")
Keta = plt.plot(Ketapril["Timepoint"],Ketapril["Survival_Percent"],color="green",marker="s",label ="Ketapril")
plac = plt.plot(Placebo["Timepoint"],Placebo["Survival_Percent"],color="black",marker="d",label ="Placebo")

#Setting legend on best place in graph
plt.legend(loc="best")

plt.title("Survival During Treatment")
plt.xlabel("Treatment Duration, Days")
plt.ylabel("Survival Rates, %")

# Save the Figure
plt.savefig("../Images/Survival.png")

# Show the Figure
plt.show()
```
